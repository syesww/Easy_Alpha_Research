```python
from rqdatac import *
import pandas as pd
init()
# 使用米筐的本地化API接口来下载数据
```

```python
start_date = '2017-01-03'
end_date = '2018-01-03'
stocks = index_components('000001.XSHG', date=start_date)
```

```python
# 将Factor和Price两个数据的获取封装成函数
def get_clean_factor(stocks, column, start_date=None, end_date=None):
    # 获取指定日期范围内的因子数据
    factor = get_factor(stocks, column, start_date=start_date, end_date=end_date)
    
    # 重置索引并按日期和证券代码排序
    factor = factor.reset_index()
    factor = factor.sort_values(['date', 'order_book_id'])
    
    # 设置日期和证券代码为索引，并删除缺失值
    factor = factor.set_index(['date', 'order_book_id']).dropna()
    
    return factor

def get_clean_price(stocks, start_date=None, end_date=None, fields='close'):
    # 获取指定日期范围内的价格数据
    price = get_price(stocks, start_date=start_date, end_date=end_date, fields=fields)
    
    # 重置索引并进行数据透视
    price = price.reset_index()
    price = price.pivot(index='date', columns='order_book_id', values=fields)
    
    return price
```

```python
price = get_clean_price(stocks, start_date, end_date)
factor = get_clean_factor(stocks, 'return_on_equity', start_date, end_date)

# 以HDF5来保存数据
current_dir = os.getcwd()
price.to_hdf(file_path, key='price', mode='w')
factor.to_hdf(file_path, key='price', mode='w')
```
